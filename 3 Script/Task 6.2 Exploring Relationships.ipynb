# 3.3. Discuss what the coefficients in the plot mean in terms of the relationships between the variables

```python
# In this analysis happiness Score is independent variable and the other variables are dependent variables. Relationships between the variables are as follows:
# With “Happiness Score”: At 0.57 we see a medium-strong relationship with "Freedom (Life Choices)". This could be interpreted as follows: the more freedom a population is, the happier they will be.
# With "Economy (GDP per Capita)": At 0.74 we see a strong relationship with "Health (Life Expectancy)". This could be interpreted as follows: As life expectancy within a nation continues to rise, so will its economy.
# With "Family (Social Support)": At 0.62 we see a strong relationship with Health (Life Expectancy). This could be interpreted as follows: As the stronger relationship a population has with their family, we can expect to see an increase in life expectancy of individuals.	
# With "Freedom "(Life Choices)": At 0.19 we see a weak relationship between Generosity (Donations to Charity). This could be interpreted: As the less a population has the freedom to make choices, the less they will be generous in donations to charities.
# With "Trust (Government Corruption)": At 0.28 we see a weak relationship between Health (Life Expectancy). This could be interpreted: As a population has less trust in its government then life expectancy will also decrease.
```

# 6. Creating a Categorical Plo

```python
# Creating a Categorical plot 
sns.histplot(df['Happiness Score'], bins = 20, kde = True)
```

```python
# Creating a Happiness flag for Countries
df.loc[df['Happiness Score'] < 4, 'Happiness Status'] = 'Unhappy Country'
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] < 6), 'Happiness Status'] = 'Somewhat Happy Country'
df.loc[df['Happiness Score'] >= 6, 'Happiness Status'] = 'Happy Country'
```

```python
# Checking creation of Exclusion flag 
df['Happiness Status'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Creating a categorical plot in seaborn using the Happiness Status created above

sns.set(style="ticks")
g= sns.catplot(x="Country", y="Happiness Score", hue="Happiness Status", data=df)
```

```python
# Majority of countries are in Somewhat happy country status as per this categorial plot and it has happiness score between 4 to 6. Then comes the happy country which has happiness score between 6 to 10. Very less countries are in Unhappy country status and it has happiness score between 0 to 4.
```
